## 3.3 직선의 방정식

직선의 방정식은 좌표에서 한 점에서 다른 점 사이를 잇는 해를 찾는 방정식이다.

### 3.3.1 두 점을 잇는 직선

일차 함수가 두 점 사이를 이어주는 점의 해를 찾아 준다.

$$
y = ax + b
$$

여기서 a는 **직선의 기울기**를 나타낸다. 이 기울기는 x가 변화함에 따라 y가 얼마나 변화하는 지를 나타내며 아래 같이 표현단다.
$$
\frac{y의 변화량}{x의 변화량}
$$ 

b는 y축과의 **절편**으로 y축과 직선이 교차하는 곳의 y좌표값이 된다.

예를 들어서 좌표에 두 점 (1,1)과 (5,3) 이 있어서 두 점을 그려보면 x 좌표가 2 늘어나면 y가 1 늘어 나므로 기울기 $a = \frac{1}{2}$ 이고, 절편 $b = 0.5$ 가 된는 것을 알 수 있다. 이 직선의 일차함수 는

$$
y = \frac{1}{2} x + \frac{1}{2}
$$

### 3.3.2 직교하는 두 직선

$y = a_1x + b_1$ 과 $y = a_2x + b_2$ 에서 두 직선의 **평행 조건**과 **직교 조건**은 아래 같다.

 - **평행조건**: 두 직선의 식에서 $a_1 = a_2$ 면 두 직선은 평행이 된다. 
 - **직교조건**: $a_1 \times a_2 = -1 $ 일 때 두 직선은 서로 수직으로 교차한다. 
 

그러면 좌표 (1,5)를 지나며 앞의 $y = \frac{1}{2} x + \frac{1}{2}$ 와 직교하는 직선의 식은?

직선의 기울기 a는 직교 조건에 따라

$$
\frac{1}{2} \times a = -1 \\
a = -2
$$

이 기울기 -2와 좌표 (1,5)를 $y = ax + b$ 에 대입하면

$$
5 = -2\times 1 + b \\
b = 7
$$

그러므로 (1,5)와 식 $y = \frac{1}{2} x + \frac{1}{2}$ 의 직교하는 식은

$$
y = -2x + 7
$$

이다.

## 3.4 비례식과 삼각비

1홉의 쌀로 밥을 지을 때 200ml 물을 넣는다. 다음에 2.5홉으로 밥을 지으려면 물을 얼마나 넣어야 하나요?

이런 문제는 비례식을 만들어서 해결이 가능합니다.

### 3.4.1 비례식의 성질

위 문제의 비례식은 아래 같습ㄴ디ㅏ.


$$
1:200 = 2.5:x \\
x = 200\times 2.5 \\
x = 500
$$

비례식은 내항과 외합의 곱한 값이 같다라는 성질이 있어어 이를 이용해 푼다.

또한 비례식은 비율을 분수식으로 처럼 나타낼 수 있다.


$$ 
a:b \rightarrow \frac{a}{b}
$$



### 3.4.2 선분, 내분점

어떤 두 점을 잇는 직성을 **선분** 이라하고 두 점을 지나 영원히 뻗는 선을 직선이라한다. 이런 선분을 $m:n$ 으로 나누는 점을 **내분점**이라 한다.

y 방향의 내분점은

$$
y = \frac{my_{2} + ny_{1}}{m + n}
$$

#### 중점

중점은 선분을 1:1로 내분하는 점이다. x, y 방향 내분점을 구하는 식의 m, n에 1을 대입해 보면 아래 등식이 성립한다.

$$
x = \frac{x_{1} + x_{2}}{2} \\
y = \frac{y_{1} + y_{2}}{2}
$$

시작점과 끝점의 x 좌표와 y 좌표를 각각 평균한 값이 중점의 좌표가 된다는 의미이다.

$$

### 3.4.3 삼각비와 원

직각삼각형은 직각 이외의 각도가 한 개 정해지면 삼각형의 모양이 정해 진다. 이 각도룰 $\theta$ 라고 하고 $\theta$가 같은 직각삼각형은 변의 길이는 달라도 그 모양은 같다. 이런 도형을 닮은꼴이라고 한다.


닮은꼴 대응하는 각 변의 비율은 같고 이것을 **닮은비**라고 한다.

$$
a1:a2 = b1:b2 = c1:c2
$$

a:b 같이 삼각형을 구성하는 변의 비율을 **형상비**라고 하고 이 비율은 가 $\theta$가 정해지면 자동으로 결정된다. 형상비 중에서 a:c, b:c, a:b 비율을 $\sin\theta$,$\cos\theta$,$\tan\theta$ 로 바꿔쓰고 이를 **삼각비**라고 한다.

<img src='https://img1.daumcdn.net/thumb/R720x0.q80/?scode=mtistory2&fname=http%3A%2F%2Fcfile27.uf.tistory.com%2Fimage%2F2452084951DF77F9075AF7' width='400'>

그리고 각 $\theta$ 의 값이 한 가지로 정해지면 두 변의 비율도 한 가지로 정해지므로 이렇게 정해지는 것을 **삼각함수**라고 한다.

삼각비를 이용해 원을 그릴 수 있다. 아래 그림은 반지름이 1인 원이다.

<img src='https://lh3.googleusercontent.com/proxy/wuVVC6nbt0GRkQtxRwgut3BZmXGJCUtFMnbFo1GU_6v-jqEmYjSttaQuE1fw9orNr75bC34V24Ls3MuwLSTgsuPJeuwTyRMz7Q' width='250'>

이 원의 원주 위에 있는 점 P의 좌표(x,y) 와 중심을 잇는 직선이 x축과 만드는 각도를 $\theta$ 라고 하면 아래 식이 성립한다.

$$
\cos\theta = \frac{x}{1} \space \sin\theta = \frac{y}{1}
$$

위 식을 변형해 P 좌표는 $(\cos\theta, \sin\theta)$ 로 나타낼 수 있다. 반지름이 r이라면 위 식에 1을 r로 치환하면 원주 위 점 P의 좌표는 $(r\times \cos\theta, r\times \sin\theta)$ 가 된다.

각 $\theta$에 대응하는 sin, cos, tan 값은 삼각함수표에서 조사할 수 있다.
